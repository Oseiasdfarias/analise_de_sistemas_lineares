# <center>Modificando Frequência Periodo e Fase de sinais</center>

## Modificando a frequência de um Sinal $f(t)$

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

import seaborn as sns
sns.set_theme()
```

## CONSIDERANDO O SINAL $f(t) = cos(t)$

<br>

### Aumentando a frequência do Sinal $f(t)$

Ao modificarmos o parâmetro do sinal para (2t) temos $f(t) = cos(2t)$, de modo que, o período do novo sinal se torna a metade do periodo do sinal original e sua frequência aumenta na mesma proporção inversa. como mostrado na figura abaixo:

Sabemos que a frequência $f_0$ de um sinal é o inverso do seu Periodo $T_0$, dada pela seguinte relação:
<br>

$$f_0 = \dfrac{1}{T_0}$$

OU

$$T_0 = \dfrac{1}{f_0}$$

<br>
Assim concluinos que, a medida que a frequência $f_0$ de um sinal aumenta seu Periodo $T_0$ diminui porporcionalmente e vice-versa. 

```python
f = lambda t: np.cos(t)

t = np.linspace(-2*np.pi, 2*np.pi, 300)

plt.figure(figsize=(16, 7))
plt.plot(t, f(t), label="t", linewidth=1.5)
plt.plot(t, f(2*t), label="2t", linewidth=1.5)

plt.xlabel('t', fontsize=16)
plt.ylabel('f(t)', fontsize=16)
ax = plt.gca()
ax.axhline(0, color='black', lw=1.2)
ax.axvline(0, color='black', lw=1.2)

plt.legend(fontsize=16)
plt.show()
```

### Diminuindo a frequência do Sinal $f(t)$


Se por outro lado, modificarmos o sinal para $f(t) = cos(t/2)$ o período do novo sinal se torna o dobro do periodo do sinal original e sua frequência diminui pela metade. como mostrado na figura abaixo:

```python
t = np.linspace(-2*np.pi, 2*np.pi, 300)

plt.figure(figsize=(16, 7))
plt.plot(t, f(t), label="t", linewidth=1.5)
plt.plot(t, f(t/2), label="2t", linewidth=1.5)

plt.xlabel('t', fontsize=16)
plt.ylabel('f(t/2)', fontsize=16)
ax = plt.gca()
ax.axhline(0, color='black', lw=1.2)
ax.axvline(0, color='black', lw=1.2)

plt.legend(fontsize=16)
plt.show()
```

### Modificando a Fase de um Sinal $f(t)$


Se quisermos modificarmos a Fase de um sinal $f(t) = cos(t)$ devesos somar um valor ao parâmetro do sinal da seguiinte forma $f(t) = cos(t+2)$ esse valor é dado em radiano, dessa forma, a fase do sinal será adiantado em 2 radianos, o gráfico abaixo mostra que o sinal modificado é transladado de 2 radianos no eixo do tempo:


```python
t = np.linspace(-2*np.pi, 2*np.pi, 300)

plt.figure(figsize=(16, 7))
plt.plot(t, f(t), label="t", linewidth=1.5)
plt.plot(t, f(t+2), label="2t", linewidth=1.5)

plt.xlabel('t', fontsize=16)
plt.ylabel('f(t+2)', fontsize=16)
ax = plt.gca()
ax.axhline(0, color='black', lw=1.2)
ax.axvline(0, color='black', lw=1.2)

plt.legend(fontsize=16)
plt.show()
```

Caso quisermos atrasar o sinal, devemos subtrair $f(t) = cos(t-2)$, assim teremos o sinal atrasado de 2 radianos no eixo do tempo como mostra a figura abaixo, dizemos que o novo sinal está defazado em 2 radianos em relação ao sinal original, ou ainda, que o sinal está atrasado em 2 radianos em relação ao sinal original.

```python
t = np.linspace(-2*np.pi, 2*np.pi, 300)

plt.figure(figsize=(16, 7))
plt.plot(t, f(t), label="t", linewidth=1.5)
plt.plot(t, f(t-2), label="2t", linewidth=1.5)

plt.xlabel('t', fontsize=16)
plt.ylabel('f(t-2)', fontsize=16)
ax = plt.gca()
ax.axhline(0, color='black', lw=1.2)
ax.axvline(0, color='black', lw=1.2)

plt.legend(fontsize=16)
plt.show()
```
