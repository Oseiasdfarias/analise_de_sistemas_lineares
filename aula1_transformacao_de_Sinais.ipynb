# <center>**Transformação de Sinais**</center>

Considerando o sinal $x(t) = 1-\dfrac{1}{2}t $ para:
 
 $$0\leq t\leq 2$$
 
 e $x(t) = 0 $ para:

 $$0 > t > 2$$

 <br>

Plotando o gráfico para esse sinal:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

%matplotlib inline
```

```python
t = np.linspace(0, 2, 20)
x_t = 1-1/2*t
plt.figure(figsize=(8, 5))
plt.plot(t, x_t)
plt.axhline(0, color='black', lw=1.2)
plt.axvline(0, color='black', lw=1.2)
plt.xlim(-0.1, 2)
plt.ylim(-0.1, 1.5)
plt.show()
```

<br>

### Mudança na Amplitude do Sinal

Se Multplicamos o sinal $x(t)$ por uma costante $k$, com $k$ satisfazendo a seguinte condição:

$$k > 1$$

A Amplitude o sinal será amplificada por $k$, no entanto seu Periodo $T$ não será afetado.

<br>
 

```python
def g_sinal(k=1):

  plt.figure(figsize = ((12, 6)))
  plt.subplot(1, 2, 1)
  plt.plot(t, x_t)
  plt.title("Gráfico de x(t)", fontsize = 16)
  plt.axhline(0, color='black', lw=1.2)
  plt.axvline(0, color='black', lw=1.2)

  plt.subplot(1, 2, 2)
  x_tm = k*(1-1/2*t)
  plt.title("Gráfico x(t) modificado", fontsize = 16)
  plt.plot(t, x_tm)
  plt.axhline(0, color='black', lw=1.2)
  plt.axvline(0, color='black', lw=1.2)
  plt.xlim(-0.1, 2)
  plt.ylim(-0.1, 2)

  plt.suptitle("Gráficos", fontsize = 20)
  plt.show()
```

```python
g_sinal(k=2)
```

<br>

Se a condição $0 < k > 1$ for satisteita:


A Amplitude do sinal será atenuada por $k$, no entanto seu Periodo $T$ não será afetado.
