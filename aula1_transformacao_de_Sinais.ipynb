# <center>**TRANSFORMAÇÕES DE SINAIS**</center>

Considerando o sinal $x(t) = cos(t)$:

 <br>

Plotando o gráfico para esse sinal:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

%matplotlib inline
```

```python
t = np.linspace(0, 2*np.pi, 100)
#x_t = 2-2*t
x_t = np.cos(t)
plt.figure(figsize=(8, 5))
plt.plot(t, x_t, linewidth=1.5)
plt.axhline(0, color='black', lw=1.2)
plt.axvline(0, color='black', lw=1.2)
plt.xlim(-1, 2*np.pi)
plt.ylim(-2, 3)
plt.show()
```

<br>

### Mudança na Amplitude do Sinal

Se Multplicamos o sinal $x(t)$ por uma costante $k$, com $k$ satisfazendo a seguinte condição:

$$k > 1$$

A Amplitude o sinal será amplificada por $k$, no entanto seu Periodo $T$ não será afetado.

<br>
 

```python
def amplitude_sinal(k=1):

  fig = plt.figure(figsize = ((15, 6)))
  ax1 = fig.add_subplot(1, 2, 1)
  ax1.tick_params(axis='x', labelsize=14)
  ax1.tick_params(axis='y', labelsize=14)
  ax1.set_xlim(-1, 2*np.pi)                    # Definindo intervalo do eixo X 
  ax1.set_ylim(-3, 3)                    # Definindo intervalo do eixo Y
  plt.plot(t, x_t, linewidth=1.5)
  ax1.set_title("Gráfico de $cos(t)$", fontsize = 16)
  plt.axhline(0, color='black', lw=1.2)
  plt.axvline(0, color='black', lw=1.2)

  # x_tm = k*(2-2*t)
  x_tm = k*np.cos(t)
  ax2 = fig.add_subplot(1, 2, 2)
  ax2.tick_params(axis='x', labelsize=14)
  ax2.tick_params(axis='y', labelsize=14)
  if (k>0):
    ax2.set_xlim(-1, 2*np.pi)                    # Definindo intervalo do eixo X 
    ax2.set_ylim(-3, 3)                    # Definindo intervalo do eixo Y
  else:
    ax2.set_xlim(-1, 2*np.pi)                    # Definindo intervalo do eixo X 
    ax2.set_ylim(-3, 3)                    # Definindo intervalo do eixo Y
  ax2.set_title(f"Gráfico de {k}$cos(t)$", fontsize = 16)
  plt.plot(t, x_tm, linewidth=1.5)
  plt.axhline(0, color='black', lw=1.2)
  plt.axvline(0, color='black', lw=1.2)

  plt.suptitle("Gráficos", fontsize = 20)
  plt.show()
```

```python
amplitude_sinal(k=2)
```

<br>

Se a condição $0 < k > 1$ for satisteita:


A Amplitude do sinal será atenuada por $k$, no entanto seu Periodo $T$ não será afetado.

<br>

```python
amplitude_sinal(k=0.5)
```

<br>

Se a condição $ k = -1$ for satisteita:


O Sinal será refletido no eixo $y$ como mostrado abaixo

<br>

```python
amplitude_sinal(k=-1)
```

<br>

Se a condição $ 0 > k >-1$ for satisteita:


O Sinal será refletido no eixo $y$ e a Amplitude será atenuada.

<br>

```python
amplitude_sinal(k=-0.5)
```
